## How deep did we have to dive ?

```python
def dive_2(child):
    if child.get("leaf", False) is True:
        if child['name'] == "Homo_sapiens":
            return {
                "distance" : child['distance'],
                "depth": 1
            }
        else:
            return None

    for ckey in ["child_1", "child_2"]:
        dive_dict = dive_2(child[ckey])
        if dive_dict is not None:
            return {
                "distance" : dive_dict["distance"] + child["distance"],
                "depth" : dive_dict["depth"] + 1
            }
```

## What Path did we take?

```python
def dive_3(child):
    if child.get("leaf", False) is True:
        if child['name'] == "Homo_sapiens":
            return {
                "distance" : child['distance'],
                "depth": 1,
                "1_2": [None]
            }
        else:
            return None

    for ckey in ["child_1", "child_2"]:
        dive_dict = dive_3(child[ckey])
        if dive_dict is not None:
            return {
                "distance" : dive_dict["distance"] + child["distance"],
                "depth" : dive_dict["depth"] + 1,
                "1_2": dive_dict["1_2"] + [ckey]
            }
```

```python
dive_3(tree)
```
